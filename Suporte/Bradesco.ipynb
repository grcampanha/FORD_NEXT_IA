```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

tabela = pd.read_excel("Bradesco.xlsx")

display(tabela)

last = tabela.index[-1] 

i = 0

while(i <= last):

    digito = tabela["DIGITO"][i]
    agencia = tabela["AGENCIA"] [i]
    conta = tabela["CONTA"] [i]
    nome = tabela["NOME"][i]
    tp = tabela["TP"][i]

    name, sobrenome = nome.split(" ", 1)


    navegador = webdriver.Chrome()
    navegador.get("https://banco.bradesco/html/classic/index.shtm")
    navegador.find_element(By.XPATH, '//*[@id="AGN"]').send_keys(str(agencia))
    navegador.find_element(By.XPATH, '//*[@id="CTA"]').send_keys(str(conta))
    navegador.find_element(By.XPATH, '//*[@id="DIGCTA"]').send_keys(str(digito))
    navegador.find_element(By.XPATH, '//*[@id="Form60"]/fieldset/ul/li[3]/input[1]').send_keys(Keys.ENTER)

    if (str(tp) == "CORRENTE"):
        p = navegador.find_elements(By.TAG_NAME ,'h2')[0].text
        p, g = p.split(" ", 2)
        
    elif(str(tp) == "SALARIO"):
    
        c = navegador.find_elements(By.TAG_NAME ,'b')[0].text

    else: 
        print("NÃO EXISTE")
        
    

    if ((str(tp) == "CORRENTE") and (name ==  g)):
        print(name, g)
        tabela.loc[tabela["NOME"] == nome, "VALIDACAO"] = str("CORRETO")
        display(tabela)
        
    elif (c == "A Conta Salário não encontra-se disponível para acesso aos Serviços do Bradesco Internet Banking."):
            print(name)
            tabela.loc[tabela["NOME"] == nome, "VALIDACAO"] = str("CORRETO")
            display(tabela)
             
        
    i = i+1   
else: 
    print(i)
```
